```python
#Import the required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.keys import Keys
```

```python
#Connecting to the web driver
driver=webdriver.Chrome('chromedriver.exe')
```

```python
#Getting webpage url
url='https://www.yatra.com/'
driver.get(url)
```

```python
# Finding web element location of search flight
search_flight=driver.find_element_by_id('BE_flight_flsearch_btn')
```

```python
search_flight
```

```python
#Clicking on the button
search_flight.click()
```

```python
Name=[]
source=[]
destination=[]
dep_time=[]
Arr_time=[]
arr_time=[]
Total_stops=[]
Duration=[]
Date_of_Journey=[]
Price=[]
```

```python
#Scrolling through the page
i=0
while(i<20):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1.2)
    i=i+1
    if i==20:
        break
        
#Scraping Airline name
airline_name=driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']")
len(airline_name)

for i in airline_name:
    Name.append(i.text)

#Scraping Source and Destination
source_destination=driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']")
len(source)

#Source

for i in range(0,len(source_destination),2):
    source.append(source_destination[i].text)

#Destination

for i in range(1,len(source_destination),2):
    destination.append(source_destination[i].text)

#Departure time
dep=driver.find_elements_by_xpath("//div[@class='i-b pr']")
for i in dep:
    dep_time.append(i.text)

#Arrival time
arr=driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']")

for i in arr:
    arr_time.append(i.text)

sep = '\n'

for i in arr_time:
    stripped = i.split(sep, 1)[0]
    Arr_time.append(stripped)

#Total Stops
stop=driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']")

for i in stop:
    Total_stops.append(i.text)
len(Total_stops)

#Duration
dur=driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")

for i in dur:
    Duration.append(i.text)
len(Duration)

#Target Variable Flight Price
price=driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")

for i in price:
    Price.append(i.text)
len(Price)

date=driver.find_element_by_xpath("//div[@class='day-li text-center cursor-pointer pr active font-primary-color']//p")

for i in range(0,len(price)):
    Date_of_Journey.append(date.text)

```
